```python
import gymnasium as gym
import time
```

```python
env = gym.make("MountainCar-v0", render_mode="human", goal_velocity=0.1)
env
```

```python
def rl_climb(observation):

    position, velocity = observation

    # go right ---->
    if -0.4 < position < 0.4:
        return 2
    
    # go left <----
    elif velocity < 0 and -0.9 < position < -0.4:
        return 0
    
    elif 0.3 < position < 0.5 or position == 0.5:
        return 1
    
    else:
        return 1
```

```python
observation,_ = env.reset()

episode_over = False
total_reward = 0

while not episode_over:
    
    action = rl_climb(observation)
    observation, reward, terminated, truncated, info = env.step(action)

    print(f"Position: {observation[0]}, Velocity: {observation[1]}")
    time.sleep(0.01)

    
    total_reward += reward
    episode_over = terminated or truncated

print(f"Total reward: {total_reward}")

env.close()    
```
